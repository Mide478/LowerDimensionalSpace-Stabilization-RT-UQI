```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Results.csv')
df.head()
```

```python
plt.subplot(121)
plt.scatter(df['N'],df['DR miu'])
plt.xlabel('sample size', size=14)
plt.ylabel('Distance ratio mean', size=14)

plt.subplot(122)
plt.scatter(df['N'],df['DR std'])
plt.xlabel('sample size', size=14)
plt.ylabel('Distance ratio standard deviation', size=14)

plt.subplots_adjust(left=0.0, bottom=0.0, right=2., top=1., wspace=0.3, hspace=0.3, )
plt.show()
```

```python
plt.subplot(121)
plt.scatter(df['N'],df['N stress'])
plt.xlabel('sample size', size=14)
plt.ylabel('stress', size=14)

plt.subplot(122)
plt.scatter(df['N+1'],df['_stress'])
plt.xlabel('OOSP', size=14)
plt.ylabel('stress', size=14)

plt.subplots_adjust(left=0.0, bottom=0.0, right=2., top=1., wspace=0.3, hspace=0.3, )
plt.show()
```

```python
plt.subplot(121)
plt.scatter(df['N'], df['stress ratio'])
plt.xlabel('sample size', size=14)
plt.ylabel('stress ratio', size=14)

plt.subplot(122)
plt.scatter(df['N'], df['rmse ratio'])
plt.xlabel('sample size', size=14)
plt.ylabel('rmse ratio', size=14)

plt.subplots_adjust(left=0.0, bottom=0.0, right=2., top=1., wspace=0.3, hspace=0.3, )
plt.show()
```

```python
plt.scatter(df['N'],df['sample ratio'])
plt.xlabel('sample size', size=12)
plt.ylabel('sample ratio', size=12)
plt.show()
```
