### Introduction

This notebook analyzes the results obtained using two different dissimilarity metrics, Euclidean and Manhattan, from the main.py script when run for experimental setups with varying sample sizes {4, 104}

Load results saved from running the main script for different ranges of sample sizes [4, 104] and OOSP's each with 100 realizations.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style("whitegrid", {'axes.grid': False})
```

```python
# Load the data from path saved with results from Euclidean Dissimilarity matrix
path = "/Results/Results with new OOSP main script/Different D's/Euclidean/"
norm_stress_N = np.load(path + "NormStress1_AllReal.npy")
norm_stress_OOSP = np.load(path + "NormStress2_AllReal.npy")
rmse_N = np.load(path + "AllRmse1_AllReal.npy")
rmse_OOSP = np.load(path + "AllRmse2_AllReal.npy")
rmse_anchor_N = np.load(path + "RmseAnchors_AllReal.npy")
rmse_anchor_OOSP = np.load(path + "RmseAnchorsExp_AllReal.npy")
```

```python
os.chdir(path)    # set the working directory to results directory
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Figure 1
ax1 = axs[0]
for i in range(norm_stress_N.shape[1]):
    row_data = norm_stress_N[:, i]
    ax1.plot(row_data, color='darkgray', alpha=0.5)
ax1.set_ylabel("Normalized Stress", fontsize=14)
ax1.set_xlabel("Sample size", fontsize=14)
ax1.set_title("N-sample case for 100 realizations", fontsize=14)

# Plot for Figure 2
ax2 = axs[1]
for i in range(norm_stress_OOSP.shape[1]):
    row_data = norm_stress_OOSP[:, i]
    ax2.plot(row_data, color='darkgray', alpha=0.5)
ax2.set_ylabel("Normalized Stress", fontsize=14)
ax2.set_xlabel("Sample size", fontsize=14)
ax2.set_title("N+1 sample case for 100 realizations", fontsize=14)

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig('Normalized Kruskal Stress for N and N+1 Cases.tiff', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Load the data from path saved with results from Manhattan Dissimilarity matrix
path = "/Results/Results with new OOSP main script/Different D's/Manhattan/"
norm_stress_N2 = np.load(path + "NormStress1_AllReal.npy")
norm_stress_OOSP2 = np.load(path + "NormStress2_AllReal.npy")
rmse_N = np.load(path + "AllRmse1_AllReal.npy")
rmse_OOSP = np.load(path + "AllRmse2_AllReal.npy")
rmse_anchor_N = np.load(path + "RmseAnchors_AllReal.npy")
rmse_anchor_OOSP = np.load(path + "RmseAnchorsExp_AllReal.npy")
```

```python
os.chdir(path)    # set the working directory to results directory
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Figure 1
ax1 = axs[0]
for i in range(norm_stress_N2.shape[1]):
    row_data = norm_stress_N2[:, i]
    ax1.plot(row_data, color='darkgray', alpha=0.1)
ax1.set_ylabel("Normalized Stress", fontsize=14)
ax1.set_xlabel("Sample size", fontsize=14)
ax1.set_title("N-sample case for 100 realizations", fontsize=14)

# Plot for Figure 2
ax2 = axs[1]
for i in range(norm_stress_OOSP2.shape[1]):
    row_data = norm_stress_OOSP2[:, i]
    ax2.plot(row_data, color='darkgray', alpha=0.1)
ax2.set_ylabel("Normalized Stress", fontsize=14)
ax2.set_xlabel("Sample size", fontsize=14)
ax2.set_title("N+1 sample case for 100 realizations", fontsize=14)

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig('Normalized Kruskal Stress for N and N+1 Cases.tiff', dpi=300, bbox_inches='tight')
plt.show()
```

```python
stress_ratio = norm_stress_OOSP/norm_stress_N # Measure of distortion
stress_ratio2 = norm_stress_OOSP2/norm_stress_N2 # Measure of distortion

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for Figure 1
ax1 = axs[0]
for i in range(stress_ratio.shape[1]):
    row_data = stress_ratio[:,i]
    ax1.plot(row_data, color='darkgray', alpha=0.1)
    ax1.set_ylabel("Stress ratio", fontsize=14)
    ax1.set_xlabel("Sample size", fontsize=14)
    ax1.set_title("Stress ratio for 100 realizations for \nEuclidean dissimilarity matrix MDS input", fontsize=14)

# Plot for Figure 2
ax2 = axs[1]
for i in range(stress_ratio2.shape[1]):
    row_data = stress_ratio2[:,i]
    ax2.plot(row_data, color='darkgray', alpha=0.1)
    ax2.set_ylabel("Stress ratio", fontsize=14)
    ax2.set_xlabel("Sample size", fontsize=14)
    ax2.set_title("Stress ratio for 100 realizations for \nManhattan dissimilarity matrix MDS input", fontsize=14)

plt.tight_layout()
plt.savefig('Stress ratio for L1 and L2 norms.tiff', dpi=300, bbox_inches='tight')
plt.show()
```
